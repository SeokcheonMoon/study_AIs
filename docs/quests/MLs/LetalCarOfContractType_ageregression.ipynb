```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', family='NanumBarunGothic') # 혹은 다른 설치한 Nanum 폰트 사용
import scipy
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import pickle
```

## 데이터

```python
df_rental = pd.read_csv("LetalCarOfContractType.csv")
df_rental
```

### 데이터 전처리 - age 결측값 제거
- target = age
- features = amount, mileage

```python
df_rental_new = df_rental.dropna(subset=['age'])
df_rental_new
```

```python
df_rental_extract = df_rental_new.loc[:, ["age","amount"]]
df_rental_extract
```

```python
df_rental_extract.isnull().sum()
```

```python
target = df_rental_extract["age"]
features = df_rental_extract[["amount"]]
target.shape, features.shape
```

## 모델

```python
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features,target,test_size = 0.1, random_state=2)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

## 평가

```python
model = LinearRegression()
```

```python
model.fit(features_train,target_train)
```

```python
# train dataset
target_train_predict = model.predict(features_train)
target_train.shape, target_train_predict.shape # 원래 정답 과 학습후 정답 수
```

```python
r2_score(target_train, target_train_predict)
```

```python
target_test_predict = model.predict(features_test)
target_test.shape, target_test_predict.shape
```

```python
r2_score(target_test, target_test_predict)
```

## 서비스

```python
with open("LetalCarOfContractType.pkl", "wb") as pickle_file :
  pickle.dump(obj=model, file=pickle_file)
```
