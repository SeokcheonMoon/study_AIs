```python
import pandas as pd
df_excel = pd.read_excel("OneWord_Game.xlsx", sheet_name=None)
type(df_excel)
```

```python
df_sheets = []
for sheet_name, df in df_excel.items():
  df_sheets.append(df)
  pass
  print(f"{sheet_name},{type(df)}")
```

```python
first = df_sheets[0].iloc[0]
second = df_sheets[1].iloc[0]
third = df_sheets[2].iloc[0]
```

```python
type(first)
```

```python
first.value_counts()
```

```python
text_concat = pd.concat(objs=(first,second,third))
df_concat = pd.DataFrame(text_concat)[1:]
text_concat
```

```python
list_values = text_concat[2:].values.tolist()         # 위에 nan값 있길래 그 다음부터 출력함.
list_values
```

```python
merge_text = ' '.join(list_values)
merge_text
```

```python
import re
patterns = "(은 |는 |이 |가 |을 |를 |들 |의 |으로 |이다. |이다 |한다. |한다 |하기 |없다. |없다 |로 |에 |에서 |\n)"
text_regex = re.sub(pattern = patterns, repl = " ", string = merge_text)
text_regex
```

```python
text_split = text_regex.split()
text_split, len(text_split)  # text의 각 요소들을 출력
```

```python
from collections import Counter
text_split_count = Counter(text_split) # text_split에서의 요소별 갯수를 확인
type(text_split_count)
```

```python
top_30 = text_split_count.most_common(30)
top_30
```

```python
answer_list ={key: value for key ,value in top_30}
answer_list
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(background_color="white", font_path = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf")
wordcloud.generate_from_frequencies(answer_list)  # 기존 text를 집어넣고 교육
```

```python
wordcloud.words_
```
